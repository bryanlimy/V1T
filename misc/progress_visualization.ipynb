```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

from sensorium.utils import tensorboard

BACKGROUND_COLOR = "#ffffff"
```

```python
font_path = "/Users/bryanlimy/Git/Lexend/Lexend-Regular.ttf"
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = prop.get_name()
```

Plot Sensorium

```python
df1 = pd.DataFrame(
    [
        ["2022-09-22", 0.12],
        ["2022-10-02", 0.37],
        ["2022-10-03", 0.39],
        ["2022-10-04", 0.42],
        ["2022-10-05", 0.45],
        ["2022-10-07", 0.46],
        ["2022-10-08", 0.45],
        ["2022-10-09", 0.54],
        ["2022-10-10", 0.55],
        ["2022-12-13", 0.56],
        ["2022-12-14", 0.576],
    ],
    columns=["Date", "Correlation"],
)
```

```python
label_fontsize, tick_fontsize = 10, 8
figure, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 3), dpi=120)
sns.scatterplot(df1, x="Date", y="Correlation", ax=ax)
ax.axhline(y=0.513, color="gray", linestyle="--", linewidth=1, alpha=0.8)
ax.text(
    x=len(df1) - 1.3, y=0.513 - 0.04, s="0.513", fontsize=tick_fontsize, color="gray"
)
tensorboard.remove_top_right_spines(axis=ax)
ax.set_xticks(
    ax.get_xticks(),
    ax.get_xticklabels(),
    rotation=45,
    ha="center",
    fontsize=tick_fontsize,
)
ax.set_ylim((0, 0.7))
ax.set_title("Sensorium progress", fontsize=label_fontsize)
x = len(df1) - 1
ax.scatter(x=x, y=0.60, color="#ff595e", s=20, alpha=0.6, label="Guanlab")
ax.scatter(x=x, y=0.598, color="#8ac926", s=20, alpha=0.6, label="Zyego")
ax.scatter(x=x, y=0.587, color="#6a4c93", s=20, alpha=0.6, label="IdV_ENS")
plt.legend(
    loc="lower center",
    frameon=False,
    ncols=3,
    framealpha=1,
    handletextpad=-0.25,
    columnspacing=1.0,
    markerscale=1,
    fontsize=tick_fontsize,
)
plt.show()
tensorboard.save_figure(
    figure, filename=os.path.join("plots", "sensorium_progress.png"), dpi=240
)
```
